#Statistics

```python
import matplotlib.pyplot as plt

def dataStatistics(df):
    print('-' * 35 + '\nintents distribution:') 
    intents_list = list(df['intent'])
    intents_freq = dict()
    for intent in set(intents_list):
        intents_freq[intent] = intents_list.count(intent)
    for key in intents_freq:
        value = intents_freq[key]
        print('%s: %d (%.2f%%),' % (key, value, value / len(intents_list) * 100))
    print()
    plt.figure(figsize=(24, 8))
    plt.bar(list(intents_freq.keys()),intents_freq.values())
    plt.xticks(rotation=90)
    plt.ylabel("frequency")
    plt.show()

    print('\n' + '-' * 35 + '\ntags distribution:')
    tags_list = list()
    for data in df['slot_method']:
        tags_list.append(data)
    tags_freq = dict()
    count = 0
    for item in tags_list:
        for tag in item:
            if (tag['slot'] in tags_freq): tags_freq[tag['slot']] += 1 
            else: tags_freq[tag['slot']] = 1
            count += 1
    for key in tags_freq:
        value = tags_freq[key]
        print('%s: %d (%.2f%%),' % (key, value, value / count * 100))
    
    print()
    plt.figure(figsize=(24, 8))
    plt.bar(list(tags_freq.keys()),tags_freq.values())
    plt.xticks(rotation=90)
    plt.ylabel("frequency")
    plt.show()
```

```python
dataStatistics(df)
```

## Train Data Statistics

```python
dataStatistics(df_train)
```

## Development Data Statistics

```python
dataStatistics(df_dev)
```

## Test Data Statistics

```python
dataStatistics(df_test)
```
